```python
import sys
import os
sys.path.insert(0, '..')
```

```python
from sna.scraping.twitter.TwitterScrap import TweetScraping
from sna.sentiment.SentimentClassifier import SentimentClassifier
```

```python
init = TweetScraping()
```

```python
init.initialize_api(wait_limit=True, rate_limit_notify=True)
```

```python
classifier = SentimentClassifier()
```

```python
data = init.search_result(query='kapal selam')
```

```python
dict_label = {0: 'netral', 1: 'positif', 2: 'negatif'}
```

```python
list_result = []
```

```python
for x in data:
    pred = classifier.predict(x['full_text'])
    x['sentiment'] = dict_label[pred]
    list_result.append(x)
```

```python
for x in list_result:
    print(x['full_text']), print(x['sentiment'])
```
